This project deals with extracting the information from a Youtube API about data related to videos on a particular channel. <br>
- It extracts data like video details, statistics of each video from the API <br>
- This data is further cleaned and put on a Dataframe for further use

```python
#import libraries
import requests
import pandas as pd
import time
```

```python
def get_video_stats(video_id):
    
    url_video_stats = "https://www.googleapis.com/youtube/v3/videos?key="+API_KEY+"&id="+video_id+"&part=statistics,id"
    response = requests.get(url_video_stats).json()

    viewCount = response['items'][0]['statistics']['viewCount']
    likeCount = response['items'][0]['statistics']['likeCount']
    #dislikeCount = response['items'][0]['statistics']['dislikeCount']
    commentCount = response['items'][0]['statistics']['commentCount']
    
    return viewCount,likeCount,commentCount
```

```python
def get_videos(df):  
    
    #Make API Call
    pageToken=""
    url_search = "https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelId="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=10000"+pageToken

    response = requests.get(url_search).json()
    time.sleep(1)

    for video in response['items']:
        if video['id']['kind'] == 'youtube#video':
            video_id = video['id']['videoId']
            video_title = video['snippet']['title']
            video_title = str(video_title).replace("&amp;","")
            upload_date = video['snippet']['publishTime']
            upload_date = upload_date.split('T')[0]
            #print("Video id is:",video_id)
        
            viewCount, likeCount, commentCount = get_video_stats(video_id)
            
            #Saving above data to a dataframe
            df = df.append({'Video ID': video_id,'Video Title': video_title,
                        'Upload Date': upload_date,'View Count': viewCount,
                       'Like Count': likeCount,'Comment Count': commentCount},ignore_index=True)
    
    return df
```

```python
#main

df = pd.DataFrame(columns=["Video ID","Video Title","Upload Date","View Count","Like Count","Comment Count"])
df = get_videos(df)
df
```
